```python
import os
import pandas as pd  
from pprint import pprint      

# 画像のパスを取得
image_dir = "./image"
image_paths = [f for f in os.listdir(image_dir) if f.endswith(".png")]
print(len(image_paths))
pprint(sorted(image_paths))

# names.csvを読み込む
names_df = pd.read_csv("./names.csv", encoding="shift_jis")
# names_df = pd.read_csv("./names.csv")
display(names_df)

# names.csvの各行に対して、画像の名前の変更を行う
for _, row in names_df.iterrows():
    old_name = row["old_name"]
    new_name = row["new_name"]

    # 画像のリストを走査して、一致する名前を探す
    for image_path in image_paths:
        # 拡張子を取り除く
        image_name_without_ext = os.path.splitext(image_path)[0]

        # 英字の大文字小文字を区別しないマッチング
        if old_name.lower() in image_name_without_ext.lower():
            # 画像名の変更
            os.rename(os.path.join(image_dir, image_path), os.path.join(image_dir, new_name + ".png"))

print("Name change process completed!")
```
